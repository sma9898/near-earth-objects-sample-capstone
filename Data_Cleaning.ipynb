```python
# Import necessary libraries
import pandas as pd
```

```python
# Load data from CSV file
df = pd.read_csv('WISE_NEACOMET_DISCOVERY_STATISTICS_rows.csv')
df.head()
```

```python
# Check for missing data
df.info()
```

```python
# Check for duplicates
df[df.duplicated()]
```

```python
# View missing values
df[df.isna().any(axis=1)]
```

```python
# There really isn't a way to calculate H, Q, or period based on the available data
# There are only 2 records missing Q and period, so it'll be easiest to drop those
df.drop(index=[147, 177], inplace=True)
```

```python
# PHA status requires BOTH MOID <= 0.05 au AND H <= 22.0. All of the remaining records with missing H have MOID > 0.05, so they would all be classified as NOT PHAs
# They're also all comets
df['PHA'] = df['PHA'].fillna('N')
```

```python
# Check for outliers
df.describe()
```

```python
# Leave H blank since they're all comets and I don't know if they have a different absolute magnitude compared to asteroids
```

```python
print(df['PHA'].value_counts())
print(df['Designation'].value_counts())
print(df['Orbit Class'].value_counts())
```

```python
# Remove * from Orbit classes
df['Orbit Class'] = df['Orbit Class'].str.replace("Jupiter-family Comet*", "Jupiter-family Comet")
df['Orbit Class'] = df['Orbit Class'].str.replace("Halley-type Comet*", "Halley-type Comet")
```

```python
# Rename columns for clarity and concision
df.rename(columns={'Discovery Date YYYY-MM-DD': 'Discovery Date', 'H (mag)': 'Magnitude', 'q (AU)': 'Perihelion', 'Q (AU)': 'Aphelion', 'i (deg)': 'Inclination', 'period (yr)': 'Orbital Period', 'MOID (AU)': 'MOID'}, inplace=True)

# Fix column data types
df['Discovery Date'] = pd.to_datetime(df['Discovery Date']) # Discovery Date as datetime
df['Orbit Class'] = df['Orbit Class'].astype('category') # Orbit Class as categorical
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
# Export as CSV
df.to_csv('NEOs_cleaned.csv')
```
